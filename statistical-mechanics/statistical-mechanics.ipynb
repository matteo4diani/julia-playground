# Introduction

Given a $3$-dimensional system of $N$ particles obeying $\alpha$ independent constraints, its degrees of freedom are $3N - \alpha$, and we can express its Hamiltonian $H(q, p)$ as a function of some coordinates $\begin{pmatrix}q_{i} \\ p_{i} \end{pmatrix}$, where $i = 1, ..., \ 3N - \alpha$.

$$
\left\{
    \begin{array}\\
        \dot{q}_{i} = \frac{\partial H}{\partial p_{i}} \\
        \dot{p}_{i} = - \frac{\partial H}{\partial q_{i}}
    \end{array}
\right.
$$

In a closed system, $H = E \ $ where $E \in \mathbb{R}$. In the fixed-energy shells that we consider in statistical mechanics this is an important assumption. 
In the real world, there's always some wiggle room for our shells: 
$$\mbox{Uncertainty}=\Delta \implies E - \Delta \leq H \leq E + \Delta$$

We are not concerned with the position and velocity of each particle: statistical mechanics aims to measure and predict macroscopic quantities as statistical realizations of microscopic processes. 

To achieve this we can express the information we have about a system in terms of a "density of hypothetical systems in the neighborhood of a given $q$ and $p$": 

$$
\rho(q,p) \ \mathrm{d^{3N}}q \ \mathrm{d^{3N}}p
$$

In many real-world cases $\rho=\rho(q,p \ ;t)$.

So, if we would like to know the number of micro-states which are compatible with a given macro-state $\{N, V, E\}$, we can compute:

$$
\Gamma(N, V, E) = \int\limits_{E - \Delta \leq H \leq E + \Delta} \! \rho(q,p \ ;t) \, \mathrm{d^{3N}}q \ \mathrm{d^{3N}}p
$$

The dependency from $N$ is in the differential element $\mathrm{d^{3N}}q \ \mathrm{d^{3N}}p$, the dependency from $V$ is obvious since we integrate on the spatial coordinates $q$, and the dependency from $E$ is in the integration domain $E - \Delta \leq H \leq E + \Delta$.

If we want to compute a function $f(q,p)$ we can use a similar reasoning to obtain:

$$
\langle f \rangle_t = \frac{\int\limits_{E - \Delta \leq H \leq E + \Delta} \! f(q,p) \ \rho(q,p \ ;t) \, \mathrm{d^{3N}}q \ \mathrm{d^{3N}}p}{\int\limits_{E - \Delta \leq H \leq E + \Delta} \! \rho(q,p \ ;t) \, \mathrm{d^{3N}}q \ \mathrm{d^{3N}}p}
$$

We are interested in stationary systems, that is systems for which:

$$
\frac{\partial \rho}{\partial t} = 0
$$


# Liouville's Theorem

The density function $\rho$ describes real (classical) physical entities that cannot disappear at will or disperse their momentum perchance, thus they must comply to the conservation of matter, assuming $\vec{v} = \left( \dot{q_i},\dot{p_i} \right)$ that is:

$$
\frac{\partial }{\partial t}\left( \int \! \rho(q,p) \, \mathrm{d}\omega \right) + \int\limits_{\sigma(\omega)} \! \rho \ \vec{v}\cdot\vec{n} \, \mathrm{d\sigma}
$$

We can transform the surface integral in the second addendum into a volume integral, per Gauss's theorem:

$$
\int\limits_{\omega} \! \left( \frac{\partial \rho}{\partial t} + \nabla\cdot\left(\rho \ \vec{v}\right) \right) \, \mathrm{d}\omega = 0
$$

Since we made no assumptions on $\omega$, the argument of the integral must be null everywhere:

$$
\frac{\partial \rho}{\partial t} + \nabla\cdot\left(\rho \ \vec{v}\right) = 0
$$

Computing the partial derivatives:

$$
\frac{\partial \rho}{\partial t} + \sum\limits_{i} \left( \frac{\partial \rho}{\partial q_i}\dot{q_i} + \rho \frac{\partial \dot{q_i}}{q_i} + \frac{\partial \rho}{\partial p_i}\dot{p_i} + \rho \frac{\partial \dot{p_i}}{p_i} \right) = 0
$$

Substituting $\dot{q_i}=\frac{\partial H}{\partial p_i}$ and $\dot{p_i}=-\frac{\partial H}{\partial q_i}$ the terms that multiply $\rho$ cancel each other:

$$
\frac{\partial \rho}{\partial t} + \sum\limits_{i} \left( \frac{\partial \rho}{\partial q_i}\frac{\partial H}{\partial p_i}  - \frac{\partial \rho}{\partial p_i}\frac{\partial H}{\partial q_i} \right) = 0
$$

Remembering the definition of Poisson's brackets:

$$
\frac{\partial \rho}{\partial t} + \left\{\rho \, ,H \right\} = 0
$$

Thus for stationary systems:

$$
\left\{\rho \, ,H \right\} = 0
$$

If we assume $\rho = \mathrm{constant} \in \mathbb{R}$ the above equation is trivially satistied and we are describing a real system that conserves matter and energy.

From now on we will express the energy shell constraint as $E \leq H \leq E + \Delta$. The simplest choice for $\rho$ is:

$$
\rho = \left\{
            \begin{array}\\
                0 & \mbox{if } \ H < E \, , H > E + \Delta \\
                const & \mbox{if } \ E \leq H \leq E + \Delta \\
            \end{array}
        \right.
$$

The product of $q$ and $p$ has the physical dimensions of an action. To find $\rho$ we can solve:

$$
\int\limits_{E \leq H \leq E + \Delta} \! (\mbox{const}) \, \mathrm{d^{3N}}q \ \mathrm{d^{3N}}p
$$

We can look for an action $h$ such that:
$$
const = \frac{1}{N! \ h^{3N}}
$$

In classical mechanics we don't have any preferential scale for actions, in quantum mechanics the scale is naturally set by Planck's constant $\hbar$. 

The $N!$ is heuristical and serves to solve Gibbs's paradox: it expresses the fact that particles are indistinguishable from a statistical (and quantum) standpoint.

Both conundrums can be seen as shortcomings in the theoretical construction of Newtonian mechanics, although they are not of much practical impediment since when we compute mean values these factors cancel themselves out:

$$
\langle f \rangle = \frac{\int \! f \ \rho \, \mathrm{d^{3N}}q \ \mathrm{d^{3N}}p}{\int \! \rho \, \mathrm{d^{3N}}q \ \mathrm{d^{3N}}p}
$$

We know $f$ is a macroscopic quantity: we can infer it from statistical mechanics or we can measure it experimentally. An experimental measure implies a mean across realizations that happen in a time interval, in the simplest case:

$$
\langle f \rangle = \frac{\int \! f \ \rho \, \mathrm{d^{3N}}q \ \mathrm{d^{3N}}p}{\int \! \rho \, \mathrm{d^{3N}}q \ \mathrm{d^{3N}}p} = \frac{1}{T} \int\limits_{0}^{T} \! \langle f \rangle \, \mathrm{d}t = \frac{1}{T} \int\limits_{0}^{T} \! f \, \mathrm{d}t = f_{exp}
$$

This is trivially provable if we assume the ergodic hypothesis.



# Microcanonical Ensemble

In a microcanonical ensemble $(N, V, E)$ are set:

$$
\Gamma = \int\limits_{E \leq H \leq E + \Delta} \! \left( \frac{1}{N! \ h^{3N}} \right) \, \mathrm{d}\omega
$$

We define entropy as:

$$
S = k_{b} \log{\Gamma}
$$

And we must now prove that such a definition of entropy obeys all the usual laws that govern entropy:

- $S$ must be extensive.
- $\frac{\partial{S}}{\partial{E}} = \frac{1}{T}$ must hold.
